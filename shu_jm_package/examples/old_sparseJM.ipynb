```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define ETF tickers and their corresponding index replacements
etf_tickers = {
    "PBUS": "mkt_index.csv",
    "VLUE": "value_index.csv",
    "SIZE": "size_index.csv",
    "MTUM": "mom_index.csv",
    "QUAL": "qual_index.csv",
    "USMV": "minvol_index.csv",
    "IWF": None  # No replacement needed
}

start_date = "1990-01-01"
end_date = "2024-01-01"

df_returns = {}
for etf, index_file in etf_tickers.items():
    try:
        data = yf.download(etf, start=start_date, end=end_date, auto_adjust=False)
        print(f"Data retrieved for {etf}:")
        print(data.head())  # Print first few rows
        print("Columns:", data.columns)  # Show available columns
        
        # Flatten MultiIndex columns if necessary
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = ['_'.join(col).strip() for col in data.columns.values]
        
        # Identify the correct 'Adj Close' column
        adj_close_col = next((col for col in data.columns if "Adj Close" in col), None)
        
        if adj_close_col:
            data[adj_close_col] = pd.to_numeric(data[adj_close_col], errors='coerce')
            df_returns[etf] = data[adj_close_col].pct_change().dropna()
        else:
            print(f"Warning: No 'Adj Close' data found for {etf}. Available columns: {data.columns}")
        
        # If the ETF has a corresponding index file and has missing earlier data, use the preloaded index data
        if index_file and index_file in dataframes:
            index_data = dataframes[index_file]
            index_data = index_data.rename(columns={"Value": etf})  # Ensure proper column naming
            index_data[etf] = pd.to_numeric(index_data[etf], errors='coerce')
            
            # Convert index data to returns
            index_returns = index_data.pct_change().dropna()
            
            # Merge ETF and index returns
            combined_returns = pd.concat([index_returns, df_returns.get(etf)], axis=0).sort_index()
            df_returns[etf] = combined_returns
    except Exception as e:
        print(f"Error downloading data for {etf}: {e}")

# Convert to DataFrame, only keeping valid columns
df_returns = pd.DataFrame({k: v for k, v in df_returns.items() if isinstance(v, pd.Series)})

df_returns = df_returns.sort_index()  # Ensure chronological order

if df_returns.empty:
    print("df_returns is empty. Printing raw data:")
    for etf, data in df_returns.items():
        print(f"{etf} Data:")
        print(data.head())
    raise ValueError("No valid return data available. Check the tickers or date range.")

# Compute cumulative returns
df_cum_returns = (1 + df_returns).cumprod()

# Plot cumulative returns
plt.figure(figsize=(12, 6))
for column in df_cum_returns.columns:
    plt.plot(df_cum_returns.index, df_cum_returns[column], label=column)
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.title("Cumulative Returns of Factor ETFs")
plt.legend()
plt.grid(True)
plt.show()
```
